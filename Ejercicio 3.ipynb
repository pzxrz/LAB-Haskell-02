```haskell
sumLista :: [Int] -> Int
sumLista xs = sum xs

mayoresQue10 :: [Int] -> Int
mayoresQue10 xs = length (filter (>10) xs)

productoLista :: [Int] -> Int
productoLista xs = product xs

celsiusAFahrenheit :: [Double] -> [Double]
celsiusAFahrenheit xs = map (\c -> c * 1.8 + 32) xs

ajustarImpares :: [Int] -> [Int]
ajustarImpares xs = map (\x -> if odd x then x + 5 else x) xs
```

```haskell
sumLista [1, 2, 3, 4]
```

```haskell
mayoresQue10 [5, 12, 8, 20]
```

```haskell
productoLista [2, 3, 4]
```

```haskell
celsiusAFahrenheit [0, 100]
```

```haskell
ajustarImpares [1, 2, 3, 4]
```

```haskell
nombresLargos :: [String] -> [String]
nombresLargos xs = filter (\nombre -> length nombre > 5) xs

soloNegativos :: [Int] -> [Int]
soloNegativos xs = filter (<0) xs

edadesUniversitarios :: [Int] -> [Int]
edadesUniversitarios xs = filter (\edad -> edad >= 18 && edad <= 25) xs
```

```haskell
nombresLargos ["Ana", "Carlos", "Esteban"]
```

```haskell
soloNegativos [5, -3, 2, -1]
```

```haskell
edadesUniversitarios [15, 18, 22, 30]
```

```haskell
agregarIVA :: [Double] -> [Double]
agregarIVA xs = map (\precio -> precio * 1.19) (filter (>=1000) xs)

longitudesVocales :: [String] -> [Int]
longitudesVocales xs = map length (filter (\palabra -> head palabra `elem` "aeiouAEIOU") xs)

cuadradosPares :: [Int] -> [Int]
cuadradosPares xs = map (^2) (filter even xs)
```

```haskell
agregarIVA [500, 1000, 2000]
```

```haskell
longitudesVocales ["oso", "casa", "elefante"]
```

```haskell
cuadradosPares [1, 2, 3, 4]
```

```haskell
import Data.Char (isUpper)

hayNegativo :: [Int] -> Bool
hayNegativo xs = any (<0) xs

todosPares :: [Int] -> Bool
todosPares xs = all even xs

nombresCorrectos :: [String] -> Bool
nombresCorrectos xs = all (\nombre -> not (null nombre) && isUpper (head nombre)) xs
```

```haskell
hayNegativo [1, 2, -3, 4]
```
