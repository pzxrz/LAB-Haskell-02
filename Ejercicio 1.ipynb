```haskell
sumLista :: [Int] -> Int
sumLista xs = sum xs

mayoresQue10 :: [Int] -> Int
mayoresQue10 xs = length (filter (>10) xs)

productoLista :: [Int] -> Int
productoLista xs = product xs

celsiusAFahrenheit :: [Double] -> [Double]
celsiusAFahrenheit xs = map (\c -> c * 1.8 + 32) xs

ajustarImpares :: [Int] -> [Int]
ajustarImpares xs = map (\x -> if odd x then x + 5 else x) xs
```

```haskell
sumLista [1, 2, 3, 4]
```

```haskell
mayoresQue10 [5, 12, 8, 20]
```

```haskell
productoLista [2, 3, 4]
```

```haskell
celsiusAFahrenheit [0, 100]
```

```haskell
ajustarImpares [1, 2, 3, 4]
```
