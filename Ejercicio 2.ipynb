```haskell
-- a. Obtener los cuadrados de los primeros 10 números
cuadrados :: [Int]
cuadrados = [x^2 | x <- [1..10]]

-- b. Filtrar los divisibles por 3 entre 1 y 30
divisiblesPor3 :: [Int]
divisiblesPor3 = [x | x <- [1..30], x `mod` 3 == 0]

-- c. Generar una lista de pares (x, y) donde x < y y ambos en [1..5]
paresOrdenados :: [(Int, Int)]
paresOrdenados = [(x, y) | x <- [1..5], y <- [1..5], x < y]
```

```haskell
cuadrados
```

```haskell
divisiblesPor3
```

```haskell
paresOrdenados
```

```haskell
divisores :: Int -> [Int]
divisores n = [x | x <- [1..n], n `mod` x == 0]
```

```haskell
divisores 12
```

```haskell
divisoresPropios :: Int -> [Int]
divisoresPropios n = [x | x <- [1..(n-1)], n `mod` x == 0]

esPerfecto :: Int -> Bool
esPerfecto n = sum (divisoresPropios n) == n
```

```haskell
divisoresPropios 6
```

```haskell
esPerfecto 6
```

```haskell
esPerfecto 10
```
