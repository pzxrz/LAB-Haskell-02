```haskell
-- Definimos los sinónimos de tipos
type Estado = (Int, Int)
type Nodo = Estado
type Costo = Int
type Grafo = [(Nodo, [(Nodo, Costo)])]

-- Capacidades de las jarras
capA, capB :: Int
capA = 5
capB = 3
```

```haskell
**PARTE 2**
```

```haskell
-- Verter de A a B
transferirAB :: Estado -> Estado
transferirAB (a, b) = 
    let cantidad = min a (capB - b)
    in (a - cantidad, b + cantidad)

-- Verter de B a A
transferirBA :: Estado -> Estado
transferirBA (a, b) = 
    let cantidad = min b (capA - a)
    in (a + cantidad, b - cantidad)
```

```haskell
transferirAB (5, 0)
```

```haskell
vecinos :: Estado -> [(Estado, Costo)]
vecinos (a, b) = 
    [ ((capA, b), 1)        -- Llenar jarra A
    , ((a, capB), 1)        -- Llenar jarra B
    , ((0, b), 1)           -- Vaciar jarra A
    , ((a, 0), 1)           -- Vaciar jarra B
    , (transferirAB (a, b), 1) -- Verter A -> B
    , (transferirBA (a, b), 1) -- Verter B -> A
    ]
```

```haskell
vecinos (0, 0)
```

```haskell
heuristica :: Estado -> Costo
heuristica (a, b) = abs (a - 4)
```

```haskell
import Data.List (sortOn)

aStar :: (Nodo -> [(Nodo, Costo)]) -> (Nodo -> Costo) -> Nodo -> Nodo -> [(Nodo, Costo)]
aStar sucesores heuristica inicio meta = buscar [((inicio, 0), 0)] []
  where
    buscar [] _ = []
    buscar (((nodo, costo), f):cola) visitados
        | nodo == meta = [(nodo, costo)]
        | nodo `elem` visitados = buscar cola visitados
        | otherwise =
            let nuevos = [ ((v, costo + c), costo + c + heuristica v) 
                           | (v, c) <- sucesores nodo, v `notElem` visitados ]
                ordenados = sortOn snd (cola ++ nuevos)
                limpios = map fst ordenados
            in (nodo, costo) : buscar limpios (nodo : visitados)
        
```
