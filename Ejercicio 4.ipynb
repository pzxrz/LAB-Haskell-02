```haskell
procesarLecturas :: (Double -> Double) -> [Double] -> [Double]
procesarLecturas f xs = map f xs
```

```haskell
procesarLecturas (\c -> c * 1.8 + 32) [0, 12, 20]
```

```haskell
crearFiltro :: (a -> Bool) -> ([a] -> [a])
crearFiltro condicion = \lista -> filter condicion lista
```

```haskell
let soloPares = crearFiltro even
```

```haskell
soloPares [1..10]
```

```haskell
enRango :: Int -> Int -> (Int -> Bool)
enRango min max = \x -> x >= min && x <= max
```

```haskell
let entre5y10 = enRango 5 10
```
